```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
train_data = pd.read_csv(r"D:\LULC\raw_data\0_groundtruth\SACLAJ_2014_2016_catclearned.csv")
test_data = pd.read_csv(r"D:\LULC\raw_data\0_groundtruth\test_points18.03.csv")
figure_path = r"C:\Users\GE\Dropbox\Kairo\under90_results\train_valin_figure"
```

```python
train_lat = train_data["target_lat"]
train_lon = train_data["target_lon"]
```

```python
test_lat = test_data["target_lat"]
test_lon = test_data["target_lon"]
```

```python
fig = plt.figure()
plt.scatter(train_lon, train_lat, s=1 )
plt.xlim(120, 155)
plt.ylim(20, 50)
plt.title("Train Data")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
fig.savefig(os.path.join(figure_path, "Train_Data.png"))
```
