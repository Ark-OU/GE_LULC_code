```python
# やったこと
# ・一枚の画像に対して一枚のマスクをかけることで欠損を補充
#    +-- 季節ごとの一枚の画像に対して同一シーン複数枚で欠損補充

# エラー
# 同じシーンでもサイズが異なっているものは使えていない

import os
from glob import glob
import matplotlib.pyplot as plt
from matplotlib import patches as mpatches
from matplotlib.colors import ListedColormap
import seaborn as sns
import numpy as np
import numpy.ma as ma
import pandas as pd
from shapely.geometry import mapping, box
import geopandas as gpd
import rasterio as rio
from rasterio.plot import plotting_extent
import earthpy as et
import earthpy.spatial as es
import earthpy.plot as ep
import earthpy.mask as em
from tqdm import tqdm
```

```python
# error_paths: use_paths[num1]に対してuse_paths[num2]の.shapeが異なっているときのpath
season_ids = ["0[1-3]", "0[4-6]", "0[7-9]", "1[0-2]"]
num1 = 0

# Main
for season_id in season_ids:
    dir_list = glob(f"E:/LULC/raw_data/1_landsat/2_landsat8_open/*20??{season_id}*-SC*")
    dir_base_list = []
    if season_id=="0[1-3]":
        season = "winter"
    elif season_id=="0[4-6]":
        season = "spring"
    elif season_id=="0[7-9]":
        season = "summer"
    elif season_id=="1[0-2]":
        season = "autumn"
    
#   Get unique id
    for i in range(len(dir_list)):
        dir_base_list.append(dir_list[i].split("\\")[1][4:10])
    scene_list = np.unique(dir_base_list)
    num_seasonfiles = len(dir_base_list)
#     print(f"{season}のシーンファイルは{num_seasonfiles}個です。")
#     winterのシーンファイルは255個です。
#     springのシーンファイルは282個です。
#     summerのシーンファイルは360個です。
#     autumnのシーンファイルは343個です。
    
#   Get season path
    cnt = 0
    season_errors = []
    
    for scene in scene_list:
        scene_paths = glob(f"E:/LULC/raw_data/1_landsat/2_landsat8_open/LC08{scene}20??{season_id}*-SC*")
        scene_paths.sort()
        
        landsat_paths_pre = glob(scene_paths[num1]+"/*band*.tif")
        landsat_paths_pre.sort()

        landsat_pre_cloud, landsat_pre_meta = es.stack(landsat_paths_pre, nodata=-9999)

        qa_path = glob(f"E:/LULC/raw_data/1_landsat/3_landsat8_use/LC08{scene}20??{season_id}*-SC*" + "/cloud_mask.tif")
        qa_path.sort()

        with rio.open(qa_path[num1]) as cloud_mask:
            landsat_qa = cloud_mask.read(1)
            landsat_pre_cloud_ext_bds  = cloud_mask.bounds

        cloud_masked = em.mask_pixels(landsat_pre_cloud, landsat_qa, vals=0)
        mask_num1 = cloud_masked.mask
        cloud_masked_copy = np.copy(cloud_masked)
        num1_shape = mask_num1.shape
        
        use_paths = glob(f"E:/LULC/raw_data/1_landsat/3_landsat8_use/LC08{scene}20??{season_id}*-SC*")
        use_paths.sort()
        landsat_paths_pre_cloud_free = glob(use_paths[num1]+"/*band*.tif")
        landsat_paths_pre_cloud_free.sort()
        
        
        # num1番目のファイルを親にして，欠損地をnum2番目のファイルで補填する．
        for num2 in range(1, len(scene_paths)):

            # Read in the "cloud free" landsat data that you downloaded as a part of your homework
            landsat_paths_pre_cloud_free = glob(use_paths[num2]+"/*band*.tif")
            landsat_paths_pre_cloud_free.sort()

            # Stack the data
            landsat_pre_cloud_free, landsat_pre_cloud_free_meta = es.stack(
                landsat_paths_pre_cloud_free,
                nodata=-9999)

            # cloud mask of num2 st data
            qa_path = glob(use_paths[num2] + "/cloud_mask.tif")
            with rio.open(qa_path[0]) as cloud_mask:
                mask_num2 = cloud_mask.read(1)

            mask_num2 = em.mask_pixels(landsat_pre_cloud_free, mask_num2, vals=0).mask
            num2_shape = mask_num2.shape
            if num1_shape != num2_shape:
                season_errors.append(use_paths[num2])
#                 print(f"scene={scene}で， num2={num2} の時に，形が違います．   num2==1~{len(scene_paths)-1} の範囲")
#                 print(f"     num1_shape={num1_shape}, num2_shape={num2_shape}")
#                 print("======================================================")
                continue
            
            # Calculate bounds - this is just for comparison later, not required
            landsat_no_clouds_bds = rio.transform.array_bounds(
                landsat_pre_cloud_free.shape[1],
                landsat_pre_cloud_free.shape[2],
                landsat_pre_cloud_free_meta["transform"])

            # Fill clouded pixels by num2 data
            cloud_masked_copy[mask_num1] = landsat_pre_cloud_free[mask_num1]

            # Update cloud mask
            mask_num1 *= mask_num2
        

        landsat_cloud_free_out_path = "E:/LULC/raw_data/1_landsat/4_landsat8_masked/"+season+"/LC08"+scene 
#         print(landsat_cloud_free_out_path)
   
        if not os.path.exists(landsat_cloud_free_out_path):
            os.makedirs(landsat_cloud_free_out_path)
            
#         landsat_clouds_clip = es.extent_to_json(list(landsat_pre_cloud_ext_bds))
#         cropped_cloud_list = es.crop_all(landsat_paths_pre_cloud_free, 
#                                          landsat_cloud_free_out_path, 
#                                          [landsat_clouds_clip], overwrite=True)

#         landsat_pre_cloud_free, landsat_pre_clod_free_meta = es.stack(
#             cropped_cloud_list)
        
        cnt += 1
        print(f"{season}の {cnt}/{num_seasonfiles} が終わりました．")
    
    print(f"{season}では{len(season_errors)}個のシーンが形に一致しませんでした.")
    if season_id=="0[1-3]":
        winter_errors = season_errors
    elif season_id=="0[4-6]":
        spring_errors = season_errors
    elif season_id=="0[7-9]":
        summer_errors = season_errors
    elif season_id=="1[0-2]":
        autumn_errors = season_errors

print("done")
```
